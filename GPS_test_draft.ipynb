```python
def sort_str(st):
  lns = st.strip().split('\n')
  lns = sorted(lns)
  return lns


def parse_dic(lns):
  dic = {}
  for ln in lns:
    ps = ln.split('\t')
    #if ' ' in ps[0]:
      #ps = ln.split(' ')
    dic[ps[0]] = int(ps[1])

  return dic


def group_by_key(dic, pf):
  s = 0
  ns = {}
  its = dic.items()
  for k, v in its:
    if k.startswith(pf):
      #ns.append(k)
      ns[k] = v
      s = s + v
  return ns, s
#

def groups_sum(dic):
  tmp = ['2', '3', '4', '5', '6', '7', '8C', '8F', '8P', '8W', '9C', '9F', '9P', '9W', 'A', 'B', 'C', 'D', 'E', 'F', 'S']
  # 2, 3, 4, 5, 6, A+   their sums are equal
  # 8C, 8F, 8P, 8W      their sums are equal
  # 9C, 9F, 9P, 9W      their sums are equal
  # C+, D+, E+, F+      their sums are equal
  for pf in tmp:
    ns, s = group_by_key(dic, pf)
    print(s, '\t', ns)
  return 

def fill_default(dic):
  #
  g1 = ['2', '3', '4', '5', '6', "A+"]
  mx1 = 0
  for pf in g1:
    ns, s = group_by_key(dic, pf)
    mx1 = max(mx1, s)
  print(mx1)
  #
  tmp = ["2N", "2R11", "2R13", "2R*"]
  for it in tmp:
    if it not in dic:
      ns, s = group_by_key(dic, '2')
      dic[it] = mx1 - s
      print(it, dic[it])
      break
  #
  tmp = ["4N", "4CL+", "4CL-"]
  for it in tmp:
    if it not in dic:
      ns, s = group_by_key(dic, '4')
      dic[it] = mx1 - s
      break
  #
  tmp = ["5N", "5WF"]
  for it in tmp:
    if it not in dic:
      ns, s = group_by_key(dic, '5')
      dic[it] = mx1 - s
      break
  #
  tmp = ["6N", "6G0", "6G1"]
  for it in tmp:
    if it not in dic:
      ns, s = group_by_key(dic, '6')
      dic[it] = mx1 - s
      break
  #
  tmp = ["A+", "A-", "AN"]
  for it in tmp:
    if it not in dic:
      ns, s = group_by_key(dic, 'A')
      dic[it] = mx1 - s
      break
  #
  #
  g2 = ['C', 'D', 'E', 'F']
  mx2 = 0
  for pf in g2:
    ns, s = group_by_key(dic, pf)
    mx2 = max(mx2, s)
  print(mx2)
  #
  tmp = ["C+", "C-", "CN"]
  for it in tmp:
    if it not in dic:
      ns, s = group_by_key(dic, 'C')
      dic[it] = mx2 - s
      break
  #
  tmp = ["D+", "D-", "DN"]
  for it in tmp:
    if it not in dic:
      ns, s = group_by_key(dic, 'D')
      dic[it] = mx2 - s
      break
  #
  tmp = ["E+", "E-", "EN"]
  for it in tmp:
    if it not in dic:
      ns, s = group_by_key(dic, 'E')
      dic[it] = mx2 - s
      break
  #
  tmp = ["F+", "F-", "FN"]
  for it in tmp:
    if it not in dic:
      ns, s = group_by_key(dic, 'F')
      dic[it] = mx2 - s
      break
  if "B" not in dic:
    dic["B"] = int(mx1 * 0.869)
  return 

```
