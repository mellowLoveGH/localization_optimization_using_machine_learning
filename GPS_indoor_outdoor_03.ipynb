```python
# filtering, Inter Quartile Range, by every feature
def filtering01(df_tmp, ratio=0.1):
  df_filtering = df_tmp.copy()
  col_nms = ['number', 'numtotal', 'R', 'snrtotal']
  #
  L = len(df_filtering)
  rn = int(ratio * 0.01 * L)
  for nm in col_nms:
    df_filtering = df_filtering.sample(frac = 1)
    df_filtering = df_filtering.sort_values(by=[nm])
    L = len(df_filtering)
    ratio = ratio * 0.01
    st = rn
    ed = L - rn
    print(nm, st, ed, ed-st, max(df_filtering[nm]), min(df_filtering[nm]))
    #print(df_filtering[:st])
    #print(df_filtering[ed:])
    df_filtering = df_filtering[ st:ed ]
  df_filtering = df_filtering.sample(frac = 1)
  print()
  return df_filtering

# sampling 
def sampling01(df_tmp, n1, n2, rto=0.4):
  df_sampling = df_tmp.copy()
  sampling_in = df_sampling.loc[ df_sampling["Class"]=="indoor" ] # 
  sampling_out = df_sampling.loc[ df_sampling["Class"]=="outdoor" ] # 
  print("original size: ", sampling_in.shape, sampling_out.shape)
  # filter outliers
  sampling_in = filtering01(sampling_in, ratio=rto)
  sampling_out = filtering01(sampling_out, ratio=rto)
  print("filtering size: ", sampling_in.shape, sampling_out.shape)
  sampling_in = sampling_in.sample(n=n1, random_state=1)
  sampling_out = sampling_out.sample(n=n2, random_state=1)
  print("sampling size: ", sampling_in.shape, sampling_out.shape)
  return sampling_in, sampling_out, pd.concat([sampling_in, sampling_out])

# sampling 
def sampling02(df_tmp, n1, rto=2.5):
  df_sampling = df_tmp.copy() 
  print("original size: ", df_sampling.shape)
  # filter outliers
  df_sampling = filtering01(df_sampling, ratio=rto)
  print("filtering size: ", df_sampling.shape)
  df_sampling = df_sampling.sample(n=n1, random_state=1)
  df_sampling["Class"] = "outdoor"
  print("sampling size: ", df_sampling.shape)
  return df_sampling

################################
log_day_in, log_day_out, log_day = sampling01(log_dataxy, 5000, 5000*1)
print("------------------------------")

#
print("------------------------------")
#win, wout, walking = sampling01(log_dataxy_msr3, 200, 1000, 0.3)
walking = filtering01(log_dataxy_msr3, ratio=2.5)
print(walking.shape)
riding = filtering01(log_dataxy_msr5, ratio=2.5)
print(riding.shape)

#
print("------------------------------ training data")
riding_train = sampling02(log_dataxy_msr5, 7000)
train_collect_data_inout = pd.concat([train_collect_data_in, train_collect_data_out, riding_train]) # 
print(train_collect_data_inout.shape)
in_num = 7000
out_num = 7000
collect_in, collect_out, collect_data = sampling01(train_collect_data_inout, in_num, out_num)
collect_data = collect_data.sample(frac = 1)
print("------------------------------ testing data")
test_collect_data_inout = pd.concat([test_collect_data_in, test_collect_data_out])
print(test_collect_data_inout.shape)
in_num = 3000
out_num = 3000
test_collect_in, test_collect_out, test_collect_data = sampling01(test_collect_data_inout, in_num, out_num)
test_collect_data = test_collect_data.sample(frac = 1)
```

```python
# shuffle data & split as training and testing
import warnings
warnings.filterwarnings("ignore")

from sklearn.utils import shuffle

def shuffle_split(df_tmp, cols, ratio=0.25):
  df_shuffled = df_tmp.copy()
  df_shuffled = shuffle(df_shuffled, random_state=0)
  # set "indoor" as 1, "outdoor" as 0
  df_shuffled["Class"] = df_shuffled["Class"].apply(lambda x: 1 if x=="indoor" else 0)
  # split as "training" & "testing"
  L = len(df_tmp)
  n1 = int(L*ratio)
  train_xy = df_shuffled[:n1]
  test_xy = df_shuffled[n1:]
  # features + class/label
  X_train = train_xy[cols[:-1]]
  y_train = train_xy[cols[-1]]
  X_test = test_xy[cols[:-1]]
  y_test = test_xy[cols[-1]]
  return X_train, y_train, X_test, y_test

def just_split(df_tmp, cols, shuffle_or_not=True):
  df_shuffled = df_tmp.copy()
  if shuffle_or_not:
    df_shuffled = shuffle(df_shuffled, random_state=0)
  df_shuffled["Class"] = df_shuffled["Class"].apply(lambda x: 1 if x=="indoor" else 0)
  X = df_shuffled[cols[:-1]]
  y = df_shuffled[cols[-1]]
  return X, y, df_shuffled

def log_evaluation(y_test, y_pred):
  yt = {0:0, 1:0}
  yp = {0:0, 1:0}
  iv = 0
  ov = 0
  #
  i = 0
  L = len(y_test)
  while i<L:
    vt = y_test[i]
    vp = y_pred[i]
    #
    if vt == 0:
      yt[0] = yt[0] + 1
    else:
      yt[1] = yt[1] + 1
    if vp == 0:
      yp[0] = yp[0] + 1
    else:
      yp[1] = yp[1] + 1
    #
    if vt==vp:
      if vt==0:
        ov = ov + 1
      else:
        iv = iv + 1
    #
    i = i + 1
  #
  yt0 = round(yt[0]/L, 3)
  yt1 = round(yt[1]/L, 3)
  yp0 = round(yp[0]/L, 3)
  yp1 = round(yp[1]/L, 3)
  print("prediction by: \t reference\t\t model:")
  print("outdoor - 0: \t\t", yt0, yt[0], "\t", yp0, yp[0])
  print("indoor - 1: \t\t", yt1, yt[1], "\t", yp1, yp[1])
  print("comparison between wifi & model: ")
  print("outdoor overlapping - 0: \t", round((ov+1)/(yt[0]+1), 3), ov)
  print("indoor overlapping - 1: \t", round((iv+1)/(yt[1]+1), 3), iv)
  return yt, yp, iv, ov, L


def evaluate_inout(y_pred, test_y):
  assert len(y_pred) == len(test_y)
  c0, c1 = 0, 0
  i = 0
  while i<len(y_pred):
    if y_pred[i] == test_y[i]:
      c0 += 1
    else:
      c1 += 1
    i = i + 1
  print( "overlap: ", c0, c0/(c0+c1) )
  print( "difference: ", c1, c1/(c0+c1) )
  return c0, c1

def evaluate_s31(y_pred):
  # 
  true_out, false_in = [], []
  i = 0
  while i<len(y_pred):
    if y_pred[i] == 0:
      true_out.append( i )
    elif y_pred[i] == 1:
      false_in.append( i )
    i = i + 1
  return true_out, false_in

def extract_gps(nums, xy):
  gps_info = []
  tmp_df = []
  for i in nums:
    row = xy.iloc[i, ]
    tmp_df.append(row)
    #print( row )
    #print( row['number'], row['number'], row['number'], row['number'], row['number'], row['number'], row['number'], row['number'] )
    gps_info.append( (round(row["gpsx"], 5), round(row["gpsy"], 5), round(row["gpsh"], 5)) )
  tmp_df = pd.DataFrame(tmp_df)
  #print(tmp_df)
  return sorted(gps_info, key=lambda x: (x[0], x[1]) ), tmp_df

def format_to_visualize(gps_info):
  st = ""
  for it in gps_info:
    st = st + str(it[0]) + "," + str(it[1]) + "|"
  return st[:-1]

# all data
test_X, test_y, test_xy = just_split(log_day, cols)
test_X01, test_y01, test_xy01 = just_split(log_day_in, cols)
test_X02, test_y02, test_xy03 = just_split(log_day_out, cols)

# msr = 3 
test_X04, test_y04, test_xy04 = just_split(walking, cols, False)
print(len(test_xy04["Class"]), sum(test_xy04["Class"]), sum(test_xy04["Class"])/len(test_xy04["Class"]))
print()

# msr = 5
test_X05, test_y05, test_xy05 = just_split(riding, cols)
print(len(test_xy05["Class"]), sum(test_xy05["Class"]), sum(test_xy05["Class"])/len(test_xy05["Class"]))
print()

# collected data - training
train_X, train_y, train_xy = just_split(collect_data, cols)
# collected data - testing
tt_X, tt_y, tt_xy = just_split(test_collect_data, cols)
```

```python
def evaluate_inout_update(y_pred, test_y):
  assert len(y_pred) == len(test_y)
  # 
  true_in, false_out = [], []
  true_out, false_in = [], []
  i = 0
  while i<len(y_pred):
    if test_y[i] == 1 and y_pred[i] == 1:
      true_in.append( i )
    elif test_y[i] == 1 and y_pred[i] == 0:
      false_out.append( i )
    elif test_y[i] == 0 and y_pred[i] == 0:
      true_out.append( i )
    elif test_y[i] == 0 and y_pred[i] == 1:
      false_in.append( i )
    i = i + 1
  print( "overlap: ", len(true_in)+len(true_out), (len(true_in)+len(true_out))/len(y_pred) )
  print( "difference: ", len(false_out)+len(false_in), (len(false_out)+len(false_in))/len(y_pred) )
  return true_in, false_out, true_out, false_in

# random forest
print("random forest")
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import precision_score, recall_score, f1_score

clf = RandomForestClassifier(n_estimators=9, criterion="gini", random_state=7, max_depth=2) # criterion: entropy
clf.fit(train_X, train_y)
print("training score: ", clf.score(train_X, train_y) )
print("testing score: ", clf.score(tt_X, tt_y) )
print()
##############
print("---------------------: data testing :--------------------")
print("msr = 3 overlap score: ", clf.score(test_X04, test_y04) )
#print("msr = 5 overlap score: ", clf.score(test_X05, test_y05) )
print("msr = 5 overlap score: ", 1-sum(clf.predict(test_X05))/len(test_y05), len(test_y05)-sum(clf.predict(test_X05)), sum(clf.predict(test_X05)), len(test_y05))
y_pred = clf.predict(test_X04)
ti, fi, to, fo = evaluate_inout_update(y_pred, test_y04.to_numpy())
print( "true in: ", len(ti), "\t false in: ", len(fi))
print( "true out: ", len(to), "\t false out: ", len(fo))
print("wifi in-out: ", sum(test_y04.to_numpy()), len(test_y04.to_numpy())-sum(test_y04.to_numpy()))
```
