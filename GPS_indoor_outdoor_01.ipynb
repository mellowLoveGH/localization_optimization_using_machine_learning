```python
# read raw log data
import os

def read_raw_log(DIR="./logdata_sub/"):
  fns = os.listdir(DIR)
  tmp = []
  for fn in fns:
    if "txt" in fn:
      tmp.append( DIR+fn )
  # 
  all_lines = []
  for fn in tmp:
    f = open(fn,"r", encoding= 'unicode_escape')
    lns = f.readlines()
    all_lines = all_lines + lns
  return all_lines

all_lines = read_raw_log()
```

```python
def string_to_dic(st):
    # for "query", "info", but "result"
    st = st.strip()
    con = st[6:-1]
    lns = con.split("&")
    dic = {}
    for ln in lns:
        ps = ln.split("=")
        if len(ps)==2:
            dic[ps[0]] = ps[1]
    return dic

def parse_raw_log(ln):
  ### get different sub-strings
  i1 = ln.find("query[")
  i2 = ln.find("info[")
  i3 = ln.find("result[")
  s1 = ln[:i1]
  s2 = ln[i1:i2] # query
  s3 = ln[i2:i3] # info
  s4 = ln[i3:]
  ### split sub-string as dic
  dic2 = string_to_dic(s2) # query
  dic3 = string_to_dic(s3) # info
  # 
  gps_str = ""
  ks02 = ["gnsf", "ll", "ll_h", "ll_n", "ll_r", "s", "mb", "msr"]
  ks03 = ["inout_flag"]
  #
  for k in ks02:
    if k not in dic2:
      return ""
    if k=="ll":
      xy = dic2[k].split("|")
      gps_str = gps_str + "llx:" + xy[0] + ",lly:" + xy[1] + ","
    else:
      gps_str = gps_str + k + ":" + dic2[k] + ","
  #
  for k in ks03:
    if k not in dic3:
      return ""
    inout = "1"
    if dic3[k]=="0" or dic3[k]=="2":
      inout = "0"
    gps_str = gps_str + inout + ","
  if dic2["gnsf"]=="-1" or dic3["inout_flag"]=="-1":
    return ""
  if dic2["gnsf"]=="0":
    return ""
  #
  return gps_str[:-1]


#
# return -> float
def cal_gpsz_tmean(gpsx, gpsy, gpsh, search_table):
  gpsx = float(gpsx)
  gpsy = float(gpsy)
  k = str(int(gpsx)*100) + "," +  str(int(gpsy)*100)
  try:
    phoneh = float(search_table[k][0])
  except:
    return 0 # not found, return default value: 0
  gpsh = float(gpsh)
  return round(abs(phoneh-gpsh))

def cal_gpsz_stdz(gpsx, gpsy, gpsh, search_table):
  d = cal_gpsz_tmean(gpsx, gpsy, gpsh, search_table)
  gpsx = float(gpsx)
  gpsy = float(gpsy)
  k = str(int(gpsx)*100) + "," +  str(int(gpsy)*100)
  try:
    phoneh = float(search_table[k][1]) # ----------- 
  except:
    return -1 # not found, return default value: 1
  if phoneh<1:
    phoneh = 1 
  return round(d/phoneh)

# query
  # ll_h, GPS海拔高度
  # ll_n, GPS卫星数
  # ll_r, GPS定位半径
  # ll, llx, lly
  # gnsf

# info
  # inout_flag
  # llx, lly, 
  # llr

# result
  # indoor

c_vi,c_mi,c_op,c_hw,c_mz,c_ss,c_ot = 0,0,0,0,0,0,0

def preprocess(gps_str, search_table):
  mbs = ["samsung", "vivo", "oppo", "mi", "huawei", "meizu", "other"]
  global c_vi ,c_mi ,c_op ,c_hw ,c_mz ,c_ss ,c_ot
  ###
  ps = gps_str.split(",")
  inout = ps[-1]
  dic = {}
  for p in ps[:-1]:
    kv = p.split(":")
    dic[kv[0]] = kv[1]
  #
  mobile_type = "none"
  if "mb" in dic:
    dic["mb"] = dic["mb"].lower()
    #print(dic["mb"])
    mobile_type = dic["mb"]
    if mbs[0] in dic["mb"]:
      c_ss = c_ss + 1
    elif mbs[1] in dic["mb"]:
      c_vi = c_vi + 1
    elif mbs[2] in dic["mb"]:
      c_op = c_op + 1
    elif mbs[3] in dic["mb"]:
      c_mi = c_mi + 1
    elif mbs[4] in dic["mb"]:
      c_hw = c_hw + 1
    elif mbs[5] in dic["mb"]:
      c_mz = c_mz + 1
      #return {}
    else:
      c_ot = c_ot + 1
      #return {}
  #
  if inout == "0":
    dic["Class"] = "outdoor"
  elif inout == "1":
    dic["Class"] = "indoor"
  #
  row = {}
  #
  gpsh = float(dic['ll_h'])
  gpsx = float(dic['llx'])
  gpsy = float(dic['lly'])
  #
  ns = dic['gnsf'].split("|")
  row['number'] = int(dic['ll_n'])
  row['snrtotal'] = int(ns[7])
  row['snrave'] = int(ns[6])
  row['numtotal'] = int(ns[2]) # int(dic['ll_n'])
  row['elesnrtotal'] = int(ns[1])
  row['elesnrave'] = int(ns[0])
  row['snr20num'] = int(ns[5])
  row['snr15num'] = int(ns[4])
  row['gpsz_tmean'] = cal_gpsz_tmean(gpsx, gpsy, gpsh, search_table)
  row['gpsz_stdz'] = cal_gpsz_stdz(gpsx, gpsy, gpsh, search_table)
  row['tenTotal'] = int(ns[3])
  row['R'] = int(dic['ll_r'])
  #
  speed = float(dic["s"])
  row['speed'] = int(speed+0.5)
  row['gpsx'] = gpsx
  row['gpsy'] = gpsy
  row['gpsh'] = gpsh
  row['mb'] = mobile_type
  row['msr'] = int(dic["msr"])
  #
  row["Class"]  = dic["Class"] 
  return row

import pandas as pd

def generate_log_df(all_lines):
  L = len(all_lines)
  i = 0
  error_count = 0
  rows_list = []
  while i<L:
    ln = all_lines[i]
    try:
      gps_str = parse_raw_log(ln)
      row = preprocess(gps_str, search_table)
      row["nth"] = i
      if row['gpsz_stdz'] != -1:
        rows_list.append( row )
    except:
      error_count += 1
    i += 1
  log_df = pd.DataFrame(rows_list)
  print( "whole length: ", L, "\t dataframe shape: ", log_df.shape, "\t error_count: ", error_count )
  return log_df

log_df = generate_log_df(all_lines)
#
c_vi,c_mi,c_op,c_hw,c_mz,c_ss,c_ot # (9595, 17017, 6373, 243, 241, 0, 262731)
```

```python
import math

def parse_data(ln):
  dic = {}
  ps = ln.strip().split("&")
  for p in ps:
    tmp = p.split("=")
    if len(tmp)==2:
      dic[tmp[0]] = tmp[1]
  return dic

def process_detail(detail):
  ps = detail.split(":")
  ele = ps[0] # ele 俯仰角
  azi = ps[1] # azimuth 方位角
  prn = ps[2] # prp, PRN (pseudo-random number) for the satellite.
  snr = ps[3] # snr信噪比
  use = ps[6] # use-in-fix(是否用到该卫星), 0: 表示没有用到, 1: 表示用到
  return ele, azi, prn, snr, use

def process_detail_col(details):
  ele, azi, prn, snr, use = [], [], [], [], []
  other_prn = []
  ds = details.strip().split("|")
  for d in ds:
    d = d.strip()
    if d=="":
      continue
    tmp = process_detail(d)
    gps_prn = int(tmp[2])
    if gps_prn>=1 and gps_prn<=32:
      ele.append(tmp[0])
      azi.append(tmp[1])
      prn.append(tmp[2])
      snr.append(tmp[3])
      use.append(tmp[4])
    else:
      other_prn.append(tmp[2])
  #print(prn)
  #print(other_prn)
  return ele, azi, snr, use

# 对于扫描到的GPS列表，扫描到所有snr信噪比大于0的卫星总数
# return -> int
def cal_numtotal(snr):
  c = 0
  for v in snr:
    if float(v)>0:
      c = c + 1
  return c

# 对于扫描到的GPS列表，扫描到所有snr信噪比大于15的卫星总数
# return -> int
def cal_snr15num(snr):
  c = 0
  for v in snr:
    if float(v)>=15:
      c = c + 1
  return c

# 对于扫描到的GPS列表，扫描到所有snr信噪比大于20的卫星总数
# return -> int
def cal_snr20num(snr):
  c = 0
  for v in snr:
    if float(v)>=20:
      c = c + 1
  return c

# 对于扫描到的GPS列表，扫描到所有snr信噪比的总和
# return -> float
def cal_snrtotal(snr):
  s = 0
  for v in snr:
    s = s + float(v)
  return round(s)

# 返回平均值
# return -> float
def cal_snrave(snr):
  s = cal_snrtotal(snr)
  n = cal_numtotal(snr)
  return round(s/n)

# 对于扫描到的GPS列表，扫描到所有snr信噪比与ele俯仰角乘积的总和
# return -> float
def cal_elesnrtotal(ele, snr):
  s = 0
  i = 0 
  while i<len(ele):
    s = s + float(ele[i])*float(snr[i])
    i = i + 1
  return round(s)

# 返回对于扫描到的GPS列表，扫描到所有snr信噪比与ele俯仰角乘积的总和/所有信噪比大于0的卫星总数的平均值
# return -> float
def cal_elesnrave(ele, snr):
  s = cal_elesnrtotal(ele, snr)
  n = cal_numtotal(snr)
  return round(s/n)

# 用来记录360°中每10°是否有卫星, 将方位角整除10向上取整,并将数值记录在num中, 所有36个数组中存在大于0数组的数量
# return -> float
def cal_tenTotal(azi):
  proportion_360 = 0
  # 用来记录360°中每10°是否有卫星
  num = []
  for i in range(37):
    num.append(0)
  #
  for v in azi:
    n = int(math.ceil( float(v)/10 ))
    num[n] = num[n] + 1
  #
  i = 1
  while i<=36:
    if num[i]>0:
      proportion_360 = proportion_360 + 1
    i = i + 1
  return round(proportion_360/36*1000) # 乘1000为了保证后面3位的精度返回int值, 

# return -> int
def cal_number(use):
  c = 0
  for v in use:
    if v=="1" or v==1:
      c = c + 1
  return c

def get_one_row(ln):
  dic = parse_data(ln)
  ele, azi, snr, use = process_detail_col(dic["detail"])
  numtotal = cal_numtotal(snr) # 对于扫描到的GPS列表，扫描到所有snr信噪比大于0的卫星总数
  snr15num = cal_snr15num(snr) # 对于扫描到的GPS列表，扫描到所有snr信噪比大于等于15的卫星总数
  snr20num = cal_snr20num(snr) # 对于扫描到的GPS列表，扫描到所有snr信噪比大于等于20的卫星总数
  snrtotal = cal_snrtotal(snr) # 对于扫描到的GPS列表，扫描到所有snr信噪比的总和
  snrave = cal_snrave(snr) # 返回平均值
  elesnrtotal = cal_elesnrtotal(ele, snr) # 对于扫描到的GPS列表，扫描到所有snr信噪比与ele俯仰角乘积的总和
  elesnrave = cal_elesnrave(ele, snr) # 返回对于扫描到的GPS列表，扫描到所有snr信噪比与ele俯仰角乘积的总和/所有信噪比大于0的卫星总数的平均值
  tenTotal = cal_tenTotal(azi)
  number = cal_number(use) # 手机将卫星用于定位的数量, use-in-fix 中 1 的个数
  R = float(dic['gpsr']) # 
  R = int(R + 0.5)
  gpsz_tmean = cal_gpsz_tmean(dic['gpsx'], dic['gpsy'], dic['gpsh'], search_table)
  gpsz_stdz = cal_gpsz_stdz(dic['gpsx'], dic['gpsy'], dic['gpsh'], search_table)
  Class = "indoor"
  if dic["inout"]=="0":
    Class = "outdoor"
  #
  tmp = [number, snrtotal, snrave, numtotal, elesnrtotal, elesnrave, snr20num, snr15num, gpsz_tmean, gpsz_stdz, tenTotal, R, Class]
  return tmp

import pandas as pd

def generate_csv(filepath, search_table):
  f = open(filepath,"r")
  lns = f.readlines()
  total_lns = len(lns)
  print("whole file length: ", total_lns)
  # header for dataframe
  head_str = "number, snrtotal, snrave, numtotal, elesnrtotal, elesnrave, snr20num, snr15num, gpsz_tmean, gpsz_stdz, tenTotal, R, Class"
  cols = head_str.split(", ")
  df = pd.DataFrame(columns = cols) 
  #
  df_count = 0
  for ln in lns[:]:
    try:
      tmp = get_one_row(ln)
      df = df.append(pd.DataFrame( [tmp], columns=cols ), ignore_index = True)
    except:
      continue
  print(df.shape)
  return df, total_lns

def generate_report(fns_df):
  df = []
  i = 0
  while i<len(fns_df[:]):
    filepath = fns_df["path"].iloc[i]
    print(i, filepath)
    tmp, total = generate_csv(filepath, search_table)
    # ---
    fns_df.iloc[i, 4] = total
    if len(tmp)>0:
      #recall = test_data(forest, tmp)
      fns_df.iloc[i, 5] = len(tmp)
      #fns_df.iloc[i, 7] = round(recall*100, 2)
    i = i + 1
    # ---
    df.append(tmp)
  df = pd.concat(df)
  # shuffle
  #df = df.sample(frac = 1).reset_index(drop=True)
  print(df.shape)
  return df

print()
```
